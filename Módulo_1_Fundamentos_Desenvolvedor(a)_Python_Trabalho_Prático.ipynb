**4 - CRIE UM CODIGO DECLARANDO AS SEGUINTES VARIAVEIS DO TIPO STRING**

```python
# Variaveis do tipo string

nome = 'João da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-0'

# Mostrando as string acima
print(nome)
print(cidade)
print(cpf)
print('\n')

# Converterdo todos os caracteres das variaveis em maiusculo
nome = nome.upper()
cidade = cidade.upper()
cpf = cpf.upper()

# Mostrando as string acima
print(nome)
print(cidade)
print(cpf)
print('\n')

# Convertendo todos os caracteres das variaiveis em minusculos
nome = nome.lower()
cidade = cidade.lower()
cpf = cpf.lower()

# Mostrando as string acima
print(nome)
print(cidade)
print(cpf)
print('\n')

# Verificar se a substring "ã" está presente na string nome
if "ã" in nome:
    # Obter a posição da primeira ocorrência da substring "ã" na string nome
    posicao = nome.find("ã")
    # Exibir a posição da substring "ã" na string
    print("A posição da substring 'ã' na string nome é:", posicao)
    print('\n')
else:
    # Caso a substring "ã" não esteja presente na string
    print("A substring 'ã' não está presente na string nome.")
    print('\n')

    # Verificar se a substring "ã" está presente na string cidade
if "ã" in cidade:
    # Obter a posição da primeira ocorrência da substring "ã" na string cidade
    posicao = cidade.find("ã")
    # Exibir a posição da substring "ã" na string cidade
    print("A posição da substring 'ã' na string cidade é:", posicao)
    print('\n')
else:
    # Caso a substring "ã" não esteja presente na string cidade
    print("A substring 'ã' não está presente na string cidade.")
    print('\n')

# Verificar se a substring "ã" está presente na string
if "ã" in cpf:
    # Obter a posição da primeira ocorrência da substring "ã" na string cpf
    posicao = cpf.find("ã")
    # Exibir a posição da substring "ã" na string cpf
    print("A posição da substring 'ã' na string cpf é:", posicao)
    print('\n')
else:
    # Caso a substring "ã" não esteja presente na string cpf
    print("A substring 'ã' não está presente na string cpf.")
    print('\n')


```
